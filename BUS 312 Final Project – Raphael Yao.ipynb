```python
t = []

client = ApifyClient('apify_api_myauXDBMvzHpeCMLUPqggtY9nb4oq417LoKF')

for ig_name in df1["ig_name"]:
    try:
        run_input = { "usernames": [ig_name] }
    
        # Run the Actor and wait for it to finish
        run = client.actor("apify/instagram-profile-scraper").call(run_input=run_input)
    except:
        continue

# Fetch and print Actor results from the run's dataset (if there are any)
    for w in client.dataset(run["defaultDatasetId"]).iterate_items():
        d ={}
        try:
            d["full name"] = w.get("fullName",np.nan)
            d["ig_name"] = w["username"]
            d["posts"] = w["postsCount"]
            d["reels"] = w["highlightReelCount"]
            d["videos"]= w["igtvVideoCount"]
            d["followers"] = w["followersCount"]
            d["following"] = w["followsCount"]
        except:
            continue
        t.append(d)

```

```python
data=pd.DataFrame(t)
```
